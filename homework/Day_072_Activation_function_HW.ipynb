# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return np.maximum(0,x)

def dReLU(x):
    return np.where(x < 0, 0, 1)
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, ReLU(x),  'b', label='ReLU(x)')
plt.plot(x, dReLU(x), 'r', label='dReLU(x)')
plt.grid()
plt.legend(loc = 2)
plt.title('ReLU Function')
plt.show()
```
